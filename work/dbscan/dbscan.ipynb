```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 500, centers=4)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps = 0.5, min_samples=10)
```

```python
model.fit(X)
```

```python
set(model.labels_)
(1 if -1 in model.labels_ else 0)
```

```python
number_label = len(set(model.labels_)) - (1 if -1 in model.labels_ else 0)
```

```python
number_label
```

```python
import numpy as np
```

```python
core_samples_mask = np.zeros_like(model.labels_, dtype=bool)
core_samples_mask[model.core_sample_indices_] = True
# Plot result
 
# Black removed and is used for noise instead.
unique_labels = set(model.labels_)
colors = ['y', 'b', 'g', 'r', 'm', 'c', 'k']
print(colors)
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'
 
    class_member_mask = (model.labels_ == k)
 
    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k',
             markersize=6)
 
    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k',
             markersize=6)
 
plt.title('number of clusters: %d' % number_label)
plt.show()
```

```python
model = DBSCAN(eps = 1, min_samples=10)
model.fit(X)
number_label = len(set(model.labels_)) - (1 if -1 in model.labels_ else 0)
core_samples_mask = np.zeros_like(model.labels_, dtype=bool)
core_samples_mask[model.core_sample_indices_] = True
# Plot result
 
# Black removed and is used for noise instead.
unique_labels = set(model.labels_)
colors = ['y', 'b', 'g', 'r', 'm', 'c', 'k']
print(colors)
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'
 
    class_member_mask = (model.labels_ == k)
 
    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k',
             markersize=6)
 
    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k',
             markersize=6)
 
plt.title('number of clusters: %d' % number_label)
plt.show()
```
