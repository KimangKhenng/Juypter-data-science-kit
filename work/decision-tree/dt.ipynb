```python
# Supervised Learning, we must split data into traning and testing 
```

```python
# When do we use testing data? We use testing data in model evaluation
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = data.target
```

```python
data.drop(['target'], axis=1,inplace=True)
```

```python
X = data.values
```

```python
y = y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtf = DecisionTreeClassifier()
```

```python
dtf.fit(X_train, y_train)
```

```python
dtf.max_features_
```

```python
y_predict = dtf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

```python
# Test with entropy critation
```

```python
dtf = DecisionTreeClassifier(criterion='entropy')
dtf.fit(X_train, y_train)
```

```python
y_predict = dtf.predict(X_test)
accuracy_score(y_test, y_predict)
```
