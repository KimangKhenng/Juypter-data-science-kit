```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
school = pd.read_csv("School_Attendance_by_Student_Group.csv")
```

```python
school.head()
```

```python
categories_count = school['Category'].value_counts()
plt.figure(figsize=(14,6))
plt.bar(categories_count.index, categories_count)
plt.title('Count plot of Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
missing = school.isnull().sum()
```

```python
totalCells = np.product(school.shape)
```

```python
totalMissing = missing.sum()
```

```python
missingPercent = (totalMissing/totalCells)*100
```

```python
print("The dataset School_Attendance_by_Student_Group contains", round(missingPercent,2),"% missing values.")
```
